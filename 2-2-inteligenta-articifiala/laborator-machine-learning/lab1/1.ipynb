```python
import numpy as np

images = np.array([np.load(f"images/car_{i}.npy") for i in range(0, 9)])
```

```python
np.sum(images)
```

```python
np.sum(images, axis=(1, 2))
```

```python
np.argmax(np.sum(images, axis=(1, 2)))
```

```python
from skimage import io
mean = np.mean(images, axis=(0))
io.imshow(mean.astype(np.uint8))
io.show()
```

```python
std = np.std(images, axis=0)
```

```python
images = (images - mean) / std
```

```python
images = images[:, 200:300, 280:400]
```

```python
for image in images:
    io.imshow(image.astype(np.uint8))
    io.show()
```
